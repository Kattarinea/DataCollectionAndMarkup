# Анализ полученного датасета

Ниже представлено описание колонок датасета moscow_places.

name — название заведения;
address — адрес заведения;
category — категория заведения, например «кафе», «пиццерия» или «кофейня»;
hours — информация о днях и часах работы;
lat — широта географической точки, в которой находится заведение;
lng — долгота географической точки, в которой находится заведение;
rating — рейтинг заведения по оценкам пользователей в Яндекс Картах (высшая оценка — 5.0);
price — категория цен в заведении, например «средние», «ниже среднего», «выше среднего» и так далее;
avg_bill — строка, которая хранит среднюю стоимость заказа в виде диапазона, например:

    «Средний счёт: 1000–1500 ₽»;
    «Цена чашки капучино: 130–220 ₽»;
    «Цена бокала пива: 400–600 ₽».
    и так далее;
middle_avg_bill — число с оценкой среднего чека, которое указано только для значений из столбца avg_bill, начинающихся с подстроки «Средний счёт»:

    Если в строке указан ценовой диапазон из двух значений, в столбец войдёт медиана этих двух значений.
    Если в строке указано одно число — цена без диапазона, то в столбец войдёт это число.
    Если значения нет или оно не начинается с подстроки «Средний счёт», то в столбец ничего не войдёт.
middle_coffee_cup — число с оценкой одной чашки капучино, которое указано только для значений из столбца avg_bill, начинающихся с подстроки «Цена одной чашки капучино»:

    Если в строке указан ценовой диапазон из двух значений, в столбец войдёт медиана этих двух значений.
    Если в строке указано одно число — цена без диапазона, то в столбец войдёт это число.
    Если значения нет или оно не начинается с подстроки «Цена одной чашки капучино», то в столбец ничего не войдёт.
chain — число, выраженное 0 или 1, которое показывает, является ли заведение сетевым (для маленьких сетей могут встречаться ошибки):
    0 — заведение не является сетевым
    1 — заведение является сетевым
district — административный район, в котором находится заведение, например Центральный административный округ;
seats — количество посадочных мест.

```python
my_data = pd.read_csv('restaurants_dataset.csv')
```

```python
my_data.drop(['lat', 'lng', 'district'], axis=1, inplace=True)
```

```python
my_data['address'][8888].split(' ', 1) #Получите расширенный доступ к странице
```

```python
my_data['city'] = my_data['address'].apply(lambda x: str(x).split(',')[0].strip())
my_data['address'] = my_data['address'].apply(lambda x: str(x).split(' ', 1)[1].lstrip(', ') if len(str(x).split(' ', 1)) > 1 else '')
my_data['address'] = my_data['address'].apply(lambda x: str(x).replace('- Петербург , ', '').strip())
```

```python
my_data.sample(10)
```

Из-за того, что на сайтах порой в поле адреса были другие данные, встречаются такие строки как "Получите расширенный доступ к странице", "Добавить отзыв". Их следует убрать

```python
my_data['address'] = my_data['address'].apply(lambda x: '' if 'Получите расширенный доступ к странице' in str(x) or 'Добавить отзыв' in str(x) else x)
```

Из списка категорий заведения видно, что кафе и Кафе / новое - одно и тоже. Значит, такого рода категории можно объединить.

```python
my_data.category.unique()
```

```python
my_data['category'] = my_data['category'].apply(lambda x: 'кафе' if 'Кафе / Новое' in str(x) or 'Кафе' in str(x) else x)
my_data['category'] = my_data['category'].apply(lambda x: 'ресторан' if 'Ресторан / Новое' in str(x) or 'Ресторан' in str(x) else x)
my_data['category'] = my_data['category'].apply(lambda x: 'бар, паб' if 'Бар / Новое' in str(x) or 'Бар' in str(x) or 
                                                                        'Паб / Новое' in str(x) or 'Паб' in str(x) or 
                                                                        'Пивной ресторан / Новое' in str(x) or 'Рюмочная' in str(x) or 
                                                                        'бар,паб' in str(x) or 'Гастропаб' in str(x) else x)
```

```python
my_data.category.unique()
```

```python
my_data['city'] = my_data['city'].apply(lambda x: 'Санкт-Петербург' if 'Санкт - Петербург' in str(x) in str(x) else x)
```

Кроме самого Новосибирска получили данные по двум городам - Бердск и Искитим, которые расположены в Новосибирской области.

Такие данные, как средняя цена чека/цена, можно было бы заменить на среднее значение или на моду, но это будут слишком грубые присвоения. Поэтому лучше оставить поля пустыми. Поле chain может быть заполнено путем поиска одинаковых названий заведений.

Для проверки на наличие дублей по названию и адресу, преобразуем немного данные к общему виду.

```python
my_data.name = my_data['name'].str.lower()
my_data.address = my_data['address'].str.lower()
my_data.hours = my_data['hours'].str.lower()
my_data.category = my_data['category'].str.lower()
my_data['name'] = my_data['name'].str.replace('ё', 'е', regex=True)
my_data['address'] = my_data['address'].apply(lambda x: str(x).replace(',', ''))
my_data['address'] = my_data['address'].apply(lambda x: re.sub(r'\s+', ' ', x))
my_data['address'] = my_data['address'].apply(lambda x: str(x).replace('\u200B', ' '))
```

```python
my_data.sample(10)
```

```python
print('Дубликатов по названию и адресу:',my_data.duplicated(subset=['name', 'address']).sum())
```

Как видно, таких дублей достаточно много, но не все стоит убирать. Например, для некоторых не указан адрес, поэтому нельзя точно сказать, являются ли они дублями. Следует проверить остальные поля. Бывает так, что на разных источниках по-разному указывают категорию (булочная/кафе) для одного заведения.

```python
my_data.loc[my_data.duplicated(subset=['name', 'address'], keep = False)]
```

```python
my_data.loc[(my_data['address'] != '') & my_data.duplicated(subset=['name', 'address'], keep=False)]
```

```python
my_data = my_data.drop(index=my_data.loc[(my_data['address'] != '') & my_data.duplicated(subset=['name', 'address'], keep=False)].index )
my_data.reset_index(drop=True)
```

```python
print(str(my_data.hours.unique()))
```

```python
my_data.hours = my_data.hours.fillna('неизвестно')
```

```python
my_data.hours = my_data['hours'].apply(lambda x: re.sub(r'\s*([,-])\s*', r'\1', x))
```

```python
my_data.loc[(my_data['address'] == '') & my_data.duplicated(subset=['name', 'address', 'hours', 'avg_bill'], keep=False)]
```

```python
my_data = my_data.drop(index=my_data.loc[(my_data['address'] == '') & my_data.duplicated(subset=['name', 'address', 'hours', 'avg_bill'], keep=False)].index )
my_data.reset_index(drop=True)
```

Пришло время заполнить колонку chain. Если в нашем датасете встречаются одинаковые названия заведений (уже после того, как убрали дубли), значит заведение является сетевым и в колонке chain нужно проставить значение 1.

```python
my_data['chain'] = my_data.duplicated(subset=['name']).astype(int)
```

# Анализ заведений по показателю chain
1 - сетевое заведение, 0 - несетевое. Как видно иг графика, несетевых заведений больше в нашем датасете.

```python
my_data['is_chain'] = my_data['chain'].map({ 0:'несетевое', 1:'сетевое'})
```

```python
chain_tmp = my_data.is_chain.value_counts().reset_index()
chain_tmp = chain_tmp.rename(columns={'type': 'type', 'count': 'count'})
chain_tmp
```

```python
fig = px.bar(chain_tmp,
             x='is_chain',
             y='count',
             title='Количество заведений в датасете по типу',
             labels={'is_chain': 'Тип', 'count': 'Количество заведений'})

fig.update_traces(hovertemplate='Тип заведения: %{x}<br>Количество: %{y}')
fig.show()
```

## Анализ количества сетевых и несетевых заведений по категориям

```python
chain_type_tmp = my_data.groupby(['category','is_chain']).agg({'chain': 'count'}).reset_index()
chain_type_tmp = chain_type_tmp.rename(columns={'category': 'category', 'is_chain':'type','chain': 'count'})
chain_type_tmp
```

Сетевыми в основном являются кофейни, рестораны и кафе.

```python
fig = px.bar(chain_type_tmp, 
             x='category', 
             y= 'count', 
             title = 'Количество сетевых и несетевых заведений по категориям', 
             color='type', barmode='group')
fig.update_layout(xaxis_title = 'Категория и тип заведения',
                  yaxis_title = 'Количество заведений')
fig.update_traces(hovertemplate = 'Категория: %{x} <br>Количество: %{y}')

fig.show()
```

```python
grouped = my_data.groupby(['category', 'is_chain'])
result = pd.DataFrame({'count': grouped.size()})
result['percentage'] = (result['count'] / result.groupby(level=0)['count'].sum() * 100).round(2)
result = result.reset_index()
result = result.query('is_chain == "сетевое"').sort_values(by = 'percentage', ascending = False)
result
```

Видно, что в процентном соотношении пекарен, фастфуда, кондитерских больше сетевых. Меньше сетевых в процентном соотношении баров, караоке и банкетных залов. Кулинарию не учитываем, как наиболее "сетевую", так как их в нашей выборке очень мало.

```python
fig = px.bar(result, 
             x='category', 
             y= 'percentage', 
             title = 'Доля сетевых заведений по категориям', 
             barmode='group')
fig.update_layout(xaxis_title = 'Категория заведения',
                  yaxis_title = 'Доля заведений')
fig.update_traces(hovertemplate = 'Категория: %{x} <br>Доля: %{y}')
fig.show()
```

Какие наиболее крупные сети представлены в нашем датасете?
В топ-10 попали заведения с названием "кафе". Его рассматривать не так интересно, так как такого рода заведения нельзя назвать сетевым.

# Анализ среднего чека

```python
my_data['avg_bill'] = my_data['avg_bill'].str.lower()
```

```python
my_data['avg_bill_from'] = my_data['avg_bill'].str.extract(r'(?:(?<!до\D)(\d+))', expand=True).astype(float)
my_data['avg_bill_to'] = my_data['avg_bill'].str.extract(r'(?:до\D*(\d+)|–(\d+)|– (\d+))', expand=True).bfill(axis=1).iloc[:, 0].fillna(-1).astype(float)
```

```python
my_data.sample(20)
```

```python
bill_tmp = my_data.groupby('category')['avg_bill_from'].mean().reset_index()
bill_tmp = bill_tmp.sort_values(by='avg_bill_from', ascending=False)
bill_tmp
```

Как видно из результатов, дешевле после пары сходить в "быстрое питание", чем в "винотеку". Для некоторых категорий у нас нет цены от, поэтому получили 0.

```python
fig = px.bar(bill_tmp,
             x='category',
             y='avg_bill_from',
             title='Количество заведений в датасете по среднему чеку ОТ',
             labels={'category': 'Категория', 'avg_bill_from': 'Средний чек ОТ'})

fig.update_traces(hovertemplate='Тип заведения: %{x}<br>Количество: %{y}')
fig.show()
```

```python
bill_tmp_to = my_data.loc[my_data['avg_bill_to'] != -1].groupby('category')['avg_bill_to'].mean().reset_index()
bill_tmp_to = bill_tmp_to.sort_values(by='avg_bill_to', ascending=False)
bill_tmp_to
```

Винотека все еще самая дорогая по чеку ДО, что логично, но печально. 

```python
fig = px.bar(bill_tmp_to,
             x='category',
             y='avg_bill_to',
             title='Количество заведений в датасете по среднему чеку ДО',
             labels={'category': 'Категория', 'avg_bill_to': 'Средний чек ДО'})

fig.update_traces(hovertemplate='Тип заведения: %{x}<br>Количество: %{y}')
fig.show()
```
